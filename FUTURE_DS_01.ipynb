```python
import pandas as pd

sales_data = {
    "Order Date": ["2025-01-01","2025-01-05","2025-01-10","2025-01-15","2025-02-01",
                   "2025-02-10","2025-02-15","2025-03-01","2025-03-10","2025-03-20"],
    "Product Name": ["Laptop","Phone","Chair","Table","Printer",
                     "Laptop","Phone","Chair","Printer","Laptop"],
    "Category": ["Technology","Technology","Furniture","Furniture","Office Supplies",
                 "Technology","Technology","Furniture","Office Supplies","Technology"],
    "Region": ["South","North","East","West","South",
               "West","East","North","South","West"],
    "Sales": [50000,30000,10000,15000,12000,
              55000,32000,11000,14000,60000]
}

df = pd.DataFrame(sales_data)

df.to_csv("sales_data.csv", index=False)

print(df)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

monthly_sales = df.groupby(df['Order Date'].dt.month)['Sales'].sum()

print(monthly_sales)

monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False)

print(top_products)

top_products.plot(kind='bar')
plt.title('Top Selling Products')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

print(category_sales)

category_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title('Category Performance')
plt.ylabel('')
plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum()

print(region_sales)

region_sales.plot(kind='bar')
plt.title('Regional Performance')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```
